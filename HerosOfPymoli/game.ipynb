```python
#import file
import pandas as pd
```

```python
#pulling file from folder
data_file= "Resources/purchase_data.csv"
```

```python
#use panda to read data
data_file_df = pd.read_csv(data_file)
data_file_df.head()
```

```python
#Total Number of Players - using a dict
players = {'Total Players': data_file_df["SN"].nunique()}
pd.DataFrame(players, index = [0])
```

```python
#Number of Unique Items
unique = data_file_df["Item ID"].nunique()

#Total Number of Purchases
total_purchase= data_file_df["Purchase ID"].count()

#Average Purchase Price
price_mean=f"${round(data_file_df['Price'].mean(),2)}"

#Total Revenue
revenue=f'${round(data_file_df["Price"].sum(),2)}'

table = {'Number of Unique Items': [unique],'Average Price':[price_mean],'Number of Purchases':[total_purchase], 'Total Revenue': [revenue]}
pd.DataFrame(table)
```

```python
#Percentage and Count of Male Players
#Percentage and Count of Female Players
#Percentage and Count of Other / Non-Disclosed

gendercounts = data_file_df.loc[:, ["Gender", "SN", "Age"]]
gendercounts = gendercounts.drop_duplicates()
gendercounts=  gendercounts["Gender"].value_counts()

players=576
genderper = (gendercounts / players)*100
genderper = genderper.astype(float).map("{:,.2f}%".format)

gender= pd.DataFrame({"Total Count":gendercounts, "Percentage of Players":genderper})

gender
```

```python
#Purchasing Analysis (Gender)
#The below each broken by gender
#Purchase Count
pc= data_file_df['Gender'].value_counts()

#withduplicates 
#Average Purchase Price
app= data_file_df.groupby(['Gender']).mean()['Price'].astype(float).map("${:,.2f}".format)

#Total Purchase Value
tpv= data_file_df.groupby(['Gender']).sum()['Price'].astype(float).map("${:,.2f}".format)

#gendercounts omitted duplicates
#Average Purchase Total per Person by Gender
atpp = data_file_df.groupby(['Gender']).sum()['Price'] / gendercounts
atpp = atpp.astype(float).map("${:,.2f}".format)

genderdata = pd.DataFrame({"Purchase Count": pc, "Average Purchase Price": app, "Total Purchase Value" :tpv, "Avg Total Purchase per Person" :atpp})    
genderdata
```

```python
#The below each broken into bins of 4 years (i.e. <10, 10-14, 15-19, etc.)

#the 1st bin is up to 9 as <10, 10 is included into the 2nd bin range of 10-14.
#always list bins with next range's end number.
bins= [0,9,14,19,24,29,34,39,50]


group_label= ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]

pd.cut(data_file_df["Age"], bins, labels=group_label).head()

data_file_df["Age Group"]= pd.cut(data_file_df["Age"], bins, labels=group_label)

data_file_df.head()

#Total Players and Percentage by age 
agecounts = data_file_df.loc[:, ["SN","Age Group"]]

#drop duplicates
agecounts = agecounts.drop_duplicates()

#groupby AGE GROUP and loop SN
agecounts= agecounts.groupby(['Age Group']).nunique()['SN']


players= 576
perplayer= (agecounts/players)*100
perplayer=perplayer.astype(float).map("{:,.2f}%".format)

agedata= pd.DataFrame({"Total Count": agecounts, "Percentage of Players": perplayer})

agedata
```

```python
#Age Demographics Table

#Purchase Count
agepc= data_file_df.groupby(['Age Group']).count()['SN']

#Average Purchase Price
ageapp= data_file_df.groupby(['Age Group']).mean()['Price'].astype(float).map("${:,.2f}".format)

#Total Purchase Value
agetpv= data_file_df.groupby(['Age Group']).sum()['Price'].astype(float).map("${:,.2f}".format)

#Average Purchase Total per Person by Age Group
ageatpp= data_file_df.groupby(['Age Group']).sum()['Price'] / agecounts
ageatpp= ageatpp.astype(float).map("${:,.2f}".format)


agetable=  pd.DataFrame({"Purchase Count": agepc, "Average Purchase Price": ageapp, "Total Purchase Value": agetpv, "Avg Total Purchase per Person": ageatpp})    
agetable
```

```python
#Top Spenders

#Identify the the top 5 spenders in the game by total purchase value, then list (in a table):

                                
#SN
#Purchase Count
spenderpc= data_file_df.groupby(['SN']).count()['Price']

#Average Purchase Price
spenderapp= data_file_df.groupby(['SN']).sum()['Price'] /spenderpc
#spenderapp= spenderapp.astype(float).map("${:,.2f}".format)

#Total Purchase Value
spendertpv= data_file_df.groupby(['SN']).sum()['Price']
#.astype(float).map("${:,.2f}".format)

spendertable= pd.DataFrame({"Purchase Count": spenderpc, "Average Purchase Price": spenderapp, "Total Purchase Value": spendertpv})
spendertable= spendertable.sort_values(by=['Total Purchase Value'], ascending=False)
spendertable['Average Purchase Price']= spendertable['Average Purchase Price'].astype(float).map("${:,.2f}".format)
spendertable['Total Purchase Value']= spendertable['Total Purchase Value'].astype(float).map("${:,.2f}".format)


spendertable.head()
```

```python
#Identify the 5 most popular items by purchase count, then list (in a table):

pd.options.display.float_format = '${:,.2f}'.format

#Item ID #Item Name #Purchase Count
item_count= data_file_df.groupby(['Item ID','Item Name']).count()['Price']

#Item Price
item_price= data_file_df.groupby(['Item ID','Item Name']).mean()['Price']

#Total Purchase Value
itemtotal= data_file_df.groupby(['Item ID','Item Name']).sum()['Price']

itemtable= pd.DataFrame({"Purchase Count": item_count, "Item Price": item_price, "Total Purchase Value":itemtotal})
itemtable= itemtable.sort_values(by=['Purchase Count'], ascending=False)

itemtable.head()
```

```python
#Identify the 5 most profitable items by total purchase value, then list (in a table):
pd.options.display.float_format = '${:,.2f}'.format
#Item ID 
#Item Name
#Purchase Count
#Item Price
#Total Purchase Value
itemtable= pd.DataFrame({"Purchase Count": item_count, "Item Price": item_price, "Total Purchase Value":itemtotal})
itemtable= itemtable.sort_values(by=['Total Purchase Value'], ascending=False)

itemtable.head()
```
